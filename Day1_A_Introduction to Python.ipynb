## 2. Lists

<u>Lists</u> are variables used to store <u>multiple elements</u>. They are defined with square brackets:

```
my_list = [2, 8, 3]
print(my_list)
```

Add a new element at the end of a previously defined list with the `append` <u>method</u>:

```
my_list.append(15)
print(my_list)
```

To access an element from a list, use its <u>index</u> (position) inside brackets (the index of the first element is 0):

```
my_list[2]
```

To access a sequence (a.k.a <u>slice</u>) of elements from a list, use:

```
my_list[0:2]
```

Note that by convention the last element (index 2) is left out.</br></br>
Leaving the second index empty returns the elements until the end of the list:

```
my_list[1:]
```

Finally, skip the <i>n</i> last elements by using a negative index after the semicolon:

```
my_list[:-2]
```

## 3. Strings

<u>Strings</u> shares some similarities with lists but they are <u>immutable</u> (you cannot modify them) and they are used to store text characters. <br>
Strings are defined with quotes (single or double):

```
string1 = "some text"
string2 = 'some more text'
print(string1)
print(string2)
```

Individual text character(s) are accessed in the same way as for lists:

```
print(string1[0:4])
```

String <u>concatenation</u> is performed with the `+` <u>operator</u>:

```
print(string1 + ' and ' + string2)
```

For increased lisibility, it is recommended to use <u>f-strings</u> instead of string concatenation:

```
print(f'{string1} and {string2}')
```

f-strings are also very convenient to <u>format</u> numerical variables to strings:

```
a = 1.632125
print(f'{string1} and {a:.2f}')
```

## 4. Conditions

To conditionally execute a <u>code block</u>, use the `if` <u>statement</u> followed by a condition and a colon:

```
number = 5
if number > 0:
  print("number is positive")
else:
  print("number is null or negative")
```

The first code block is only executed if the condition is <u>True</u>.<br>
The <u>else</u> statement defines an alternate code block executed if the condition is <u>False</u>.<br>
To avoid syntax error and clearly delimitate condition blocks, it is <u>required to indent</u>!


The existing numerical comparison operators are:
- greater: '>', smaller: `<`, greater or equal: `>=`, smaller or equal: `<=`, inequal to `!=`
- the equal to operator `==` is not to be confused with the assignment operator `=`

For increased lisibility, it is also possible to negate a condition with the  statement `not`:

```
if not number == 0:
  print("number is not zero")
```

To check if an element is in a list, use the `in` statement:

```
if (5 in my_list):
  print("5 is in my_list")
else:
  print("5 is not in my_list")
```

As in plain English, you can also use `and` and `or` to combine conditions:

```
if (2 in my_list) and (3 in my_list):
  print("2 and 3 are in my_list")
```

In Python everything is ultimately an <u>object</u>, and objects are not copied but <u>referenced</u> when assigning them to a new variable:

```
a = [1, 2, 3]
b = a
b[0] = 4
print(a)
```

To check whether two objects are identical use the condition statement `is`:

```
print(a is b)
```

## 5. Loops

<u>Loops</u> are used to execute a code block multiple times; there are two types of loops in Python: `for` and `while`.

`for` loops are commonly used to <u>iterate sequentially</u> over the elements of a list:

```
for element in my_list:
  print(element)
```

Note that the variable _element_ is an <u>iteration variable</u>.<br>
It could have been called with any valid variable name and it is <u>only defined inside the loop block</u> (indented code).

It is also possible to iterate over a `range` of integers with `range`:

```
for i in range(0, 5):
  print(i)
```

Or over both the indexes and the elements of a list with `enumerate`:

```
for i, element in enumerate(my_list):
  print(i, element)
```

Finally, multiple loops can be <u>nested</u> (with correct indentation!):

```
for i in range(0, 3):
  for j in range(0, 2):
    print(i, j)
```

`While` loops are commonly used when the number of iterations to be performed is not known beforehand.<br>
Here for instance to <u>iteratively</u> compute factorials and find the smallest <u>factorial</u> greater than 5000:

```
integer = 1
factorial = 1
while factorial < 5000:
  integer = integer+1
  factorial = factorial*integer
print(f'Factorial ({integer}) = {factorial} is the smallest factorial greater than 5000')
```

## 6. Functions

To improve readability and <u>modularity</u>, a good practice is to package the code of <u>repetitive atomic tasks</u> into <u>functions</u>.<br><br>
A function can take <u>input arguments</u> and it can return an <u>output argument</u> specified by the `return ` statement:

```
def square(x):
  return x ** 2

def multiply(a, b):
  return a * b
```

To <u>call</u> a function, type its name followed by parenthesis and pass as many arguments as needed separated by commas:

```
print(multiply(3,2))
```

Functions can be <u>composed</u>:

```
print(square(multiply(3, 2)))
```

To improve code readability, function arguments can be explicitely named as in the function definition:

```
print(square(multiply(a=3, b=2)))
```

Finally, a function can be defined with <u>default arguments</u>:

```
def say_something(text='Hi there!'):
  print(text)

say_something()
say_something('Python is fun')
```

Arguments with <u>no default value</u> must come first in the function definition and they must be passed when calling the function:

```
def poweradd(a, b, power=1):
  return (a+b)**power

print(poweradd(3, 2, power=2))
```
